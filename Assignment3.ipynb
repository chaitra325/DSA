```python
#Pairplot to show scatter plots for all pairs of features
print("\n--- Visualization 1: Pairplot ---")
sns.pairplot(df,hue='Classification', height=2)
plt.suptitle('Pair Plot of Iris Features by Species', y=1.05)
plt.show()
```

```python
#Scatterplot to understand relationship between sepal diamensions
print("\n--- Visualization 2: Scatterplot---")
sns.scatterplot(x='SL', y='SW',hue='Classification', data=df, )
plt.legend(bbox_to_anchor=(1, 1), loc=2)
plt.title('Sepal Length vs. Sepal Width by Species')
plt.show()
```

```python
#Boxplot showing distribution of dataset
import warnings
warnings.filterwarnings('ignore')
print("\n--- Visualization 3: Boxplot---")
def graph(y):
    sns.boxplot(x='Classification', y=y, data=df,palette='husl')

plt.figure(figsize=(8,8))
plt.subplot(221)
graph('SL')
plt.subplot(222)
graph('SW')
plt.subplot(223)
graph('PL')
plt.subplot(224)
graph('PW')

plt.show()
```
